Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price

```python
import selenium                                  #importing the required libararies
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com/')              #connecting to the page
```

```python
product=driver.find_element(By.CLASS_NAME,'_3704LK')
product.send_keys('sunglasses')                         #to eneter "sunglasses" in the search box
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search.click()       #to click the search button
```

```python
# to scrap 100 data for sunglasses

sunglasses_brand=[]
product_description=[]
price=[]

start=0
end=3
for page in range(0,3):
 
    # to scrap the brand of sunglasses:
    brand=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand:
        sunglasses=i.text
        sunglasses_brand.append(sunglasses)
  
    # to scrap the description of the sunglasses:
    description=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in description:
        product_description.append(i.text)
        
    # to scrap the price of the sunglasses:      
    sunglass_price=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')    
    for i in sunglass_price:
        price.append(i.text)

    #to click the next button
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')    
    next_button.click()
    time.sleep(3)
    
sunglasses_brand=sunglasses_brand[0:100]
product_description=product_description[0:100]
price=price[0:100]

print(len(sunglasses_brand),len(product_description),len(price))    
```

```python
data_sunglass={"BRAND":sunglasses_brand,"PRODUCT_DESCRIPTION":product_description,"PRICE":price}
df_sunglasses=pd.DataFrame(data_sunglass)
df_sunglasses                     #making the dataFrame
```

Q8.Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpage https://www.azquotes.com/
2. Click on Top Quotes

```python
import selenium                     #importing the required libraries
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.azquotes.com/')                #to connect to the page 
```

```python
top_quotes=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes.click()                   #to click on top quotes
```

```python
#to scrap top 1000 Quotes of all time:
start=0
end=15

for page in range(0,15):   #to scrap 100 data (for scraping 1000 we can make the range 0 to 150)
    
  #to scrap the Quote:
    quote=[]
    quote_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tag:
        quote.append(i.text)
        
   #to scrap Author:
    author=[]
    author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]')   
    for i in author_tag:
        author.append(i.text)
        
   #to scrap type of quotes:
    
    quote_type=[]
    type_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]') 
    for i in type_tag:
        quote_type.append(i.text)
        
  # for next button
    
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]')    
    next_button.click()
    time.sleep(3)
        
print(len(quote),len(author),len(quote_type))        

```

```python
data_quote={"QUOTE":quote,"AUTHOR":author,"QUOTE_TYPE":quote_type}              #making the dataframe
df_quote=pd.DataFrame(data_quote)
df_quote
```
