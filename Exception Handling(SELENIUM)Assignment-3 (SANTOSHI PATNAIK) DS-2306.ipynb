Q2: from the above search reasult scrap the following details of first 3 pages :
        1.Brand name, 2.Name of the product, 3.Price, 4. return/Exchange, 5.Epected Delivery, 6. Availability, 7.Product URL

```python
driver=webdriver.Chrome()
driver.get('https://www.amazon.in/')

search_item=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search_item.send_keys("guitar")

search_button=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search_button.click()
time.sleep(5)
# fetching url to open each product   from quesion no.1(guitar)

start=0
end=3
Product_URL=[]
for page in range(start,end):
    
    p_url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in p_url:
        Product_URL.append(i.get_attribute("href"))
        
    next_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()                                 
    time.sleep(3)
Brand=[]
Product_name=[]
Price=[]
Returns=[] 
Expected_delivery=[]
Availability=[]
    

for p_url in Product_URL[0:30]:
    
                                            #loop for every guitar in the list
    driver.get(p_url)
    time.sleep(5)
    
     # for extracting brand name
    try:                                               
        brand=driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        Brand.append(brand.text)
        
    except NoSuchElementException:
         Brand.append('-')
     #for extracting the name of product
    try:
        name=driver.find_element(By.XPATH,'//*[@id="title_feature_div"]/div/h1/span')
        Product_name.append(name.text)
    except NoSuchElementException:
        Product_name.append('-')
         
     # for extracting Price:        
    try:
        p_tag=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        Price.append(p_tag.text)
        
    except NoSuchElementException:
        Price.append('-')
        
    #for scraping return/exchange:
    try:
        returns_tag=driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/span')
        Returns.append(returns_tag.text)
        
    except NoSuchElementException:
        Returns.append('-')
    
    # for scraping expected delivery:
    try:
        delivery_tag=driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK"]/div/span/span')
        Expected_delivery.append(delivery_tag.text)
    except NoSuchElementException:
        Expected_delivery.append('-')
    
    # for scraping Availability:
    try:
        a_tag=driver.find_element(By.XPATH,'//*[@id="availabilityInsideBuyBox_feature_div"]/div/div/span')
        Availability.append(a_tag.text)
    except NoSuchElementException:
        Availability.append('-')
    
    #for scraping url
    ##try:
        #url_tag=driver.find_element(By.XPATH,'')
print(len(Brand),len(Product_name),len(Price),len(Returns),len(Expected_delivery),len(Availability))
    
```

```python
try:
    
    url_string=driver.current_url
    print("URL Extracted:",url_string)
    lat_lng=re.findall(r'@(.*)data',url_string)
    if len(lat_lng):
        lat_lng_list=lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat=lat_lng_list[0]
            lng=lat_lng_list[1]
        print("Latitude={},Longitude={}".format(lat,lng))
except Exception as e:
    print("Error:",str(e))
    
```
